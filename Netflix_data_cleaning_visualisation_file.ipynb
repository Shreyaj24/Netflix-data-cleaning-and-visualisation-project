```python
import pandas as pd                 #importing pandas for data cleaning and manipulating
import numpy as np                  #importing numpy for numerical operations
import seaborn as sns               #importing seaborn for data visualisation
import matplotlib.pyplot as plt     # importing matplotlib for graphs and visualisations
import calendar                     # importing calendar for fetching month from date. 
```

# Data cleaning and Manipulation

```python
# reading the netflix file using pandas read_csv functionality
netflix_df = pd.read_csv('netflix.csv') 
netflix_df.head()                   #to display top 5 records
```

```python
netflix_df.info()
```

```python
# to deal with null values
netflix_df.isnull().sum()
```

```python
# to drop the null values for title and date_added
netflix_df.dropna(subset=['title'], inplace = True)
netflix_df.dropna(subset=['date_added'], inplace = True)
netflix_df
```

```python
# renaming the columns
netflix_df.rename(columns={"listed_in":"category"},inplace=True)
netflix_df.head()
```

```python
# to reset the index
netflix_df.reset_index(inplace=True)
netflix_df
```

```python
# to get the uniformity in the date_added column
netflix_df['date_added'] = pd.to_datetime(netflix_df['date_added'])
netflix_df
```

```python
netflix_df.type.unique()
```

```python
netflix_df.rating.unique()
```

```python
# create another column for season and movie(seasons and movie_duration)
netflix_df['seasons'] = netflix_df.loc[:, 'duration']                #copy the duration column into seasons column
netflix_df
```

```python
# changing the order of column
netflix_df = netflix_df[['index','show_id', 'type','title', 'director', 'country','date_added', 'release_year', 'rating', 'duration', 'seasons', 'category']]
netflix_df 
```

```python
# moving na value to the seasons column by filtering with type column.
netflix_df.loc[netflix_df['type'].str.contains('Movie'), 'seasons'] = 'NA'
netflix_df
```

```python
# moving na value to the seasons column by filtering with type column.
netflix_df.loc[netflix_df['type'].str.contains('TV Show'), 'duration'] = 'NA'
netflix_df
```

```python
# To remove string seasons from the seasons column in records
netflix_df['seasons'] = netflix_df['seasons'].map(lambda x: x.rstrip('Season'))
netflix_df

```

```python
# To remove min from the duration column in records
netflix_df['duration'] = netflix_df['duration'].map(lambda x: x.rstrip('min'))
netflix_df
```

```python
# To convert duration in mins if given in hours
def to_minutes(time_string):
    #print(time_string)
    if 'hr' in str(time_string) or 'hour' in str(time_string):
        hours, _, minutes,_ = time_string.split(' ')
        print(hours, minutes)
        return int(hours) * 60 + int(minutes)
    else:
        return time_string
    
netflix_df['duration'] = netflix_df['duration'].apply(to_minutes) 
netflix_df
```

```python
# Adding rating_category column for better description of the ratings

def get_rating_category(rating_value):
    if rating_value == 'PG-13':
        return 'Under 13 with parental Guidance'
    elif rating_value == 'TV-MA':
        return 'Above 17'
    elif rating_value == 'TV-PG':
        return 'Under 10 with parental guidance'
    elif rating_value == 'TV-14':
        return 'Under 14 with parental Guidance'
    elif rating_value == 'TV-Y7':
        return 'Above 7'
    elif rating_value == 'TV-Y':
        return 'Age 2 to 6'
    elif rating_value == 'PG':
        return 'Under 7 with parental guidance'
    elif rating_value == 'TV-G' or rating_value == 'G':
        return 'Genreal Audiences : suitable for all age group'
    elif rating_value == 'R':
        return 'Restricted : under 17 with parental guidance'
    elif rating_value == 'NC-17':
        return 'Adults only: No one 17 and under admitted'
    elif rating_value == 'NR'or rating_value == 'UR':
        return 'Not/Un rated'
    elif rating_value == 'TV-Y7-FV':
        return 'Directed to Older Children - Fantasy Violence'
    else:
        return 'category cannot be specified'

netflix_df['rating_category'] = netflix_df['rating'].apply(get_rating_category) 
netflix_df
```

```python
netflix_df.rename(columns={"duration":"movie_duration"},inplace=True)
netflix_df.head()
```

```python
# changing the order of column
netflix_df = netflix_df[['index','show_id', 'type','title', 'director', 'country','date_added', 'release_year', 'rating', 'rating_category', 'movie_duration', 'seasons', 'category']]
netflix_df 
```

```python
# converting release_year column to int datatype
netflix_df['release_year'] = netflix_df['release_year'].astype(str).str.split('.').str[0]
netflix_df['release_year'] = netflix_df['release_year'].astype(int)
netflix_df
```

# Data Visualisation

```python
# creating pie chart to understand movie count and TV Show count in netflix

# Define data
type_count = netflix_df['type'].value_counts()
show_labels = ['Movie', 'TV Show']
my_explode = [0, 0.1]

# Define colors for the partitions
colors = ["skyblue", "lightgreen"]

# Plot pie chart
plt.figure(figsize=(8, 8))  
plt.pie(type_count, labels=show_labels, autopct='%1.1f%%', explode=my_explode, shadow=True, colors=colors, startangle=90)

# Add title and legend
plt.title('Distribution of Shows on Netflix')
plt.legend(title='Show Type', loc='upper right')

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
# Rating count
#value_counts() returns a Series containing counts of unique values in a DataFrame column. It can be used to quickly understand the distribution of values in a categorical column.
rating_count = netflix_df['rating'].value_counts().reset_index()
rating_count
```

```python
# Create the horizontal bar chart for the representation of rating count
# create the plot
plt.figure(figsize=(10, 6))  

# Define color palette
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
rating_count.sort_index().plot(kind='barh', color= colors, edgecolor='black')

#Title and labels
plt.title('Rating Distribution on Netflix' ,fontsize = 16)
plt.xlabel('Rating', fontsize =14)
plt.ylabel('Count', fontsize =14)

# Rotate x-axis labels and add grid lines
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.tight_layout()  
plt.show()
```

```python
# Taking top 10 country count for movies and TV shows by categorising on country and type column using groupby()
country_count = netflix_df.groupby(['country', 'type']).size().reset_index(name='count')
country_count = country_count.sort_values(by='count', ascending=False).head(10)
country_count
```

```python
# Bar chart to represent top 10 countries based on the count of movies and TV shows
plt.figure(figsize=(12,8))
plt.title('Top 10 Countries with Most Movies and TV Shows on Netflix', fontsize=16)

# Using a custom color palette
colors = sns.color_palette("viridis", len(country_count))

# create the bar plot
sns.barplot(x='country',y='count',data=country_count, palette = colors, hue = country_count["type"])

plt.xlabel('Country', fontsize=14)
plt.ylabel('No of Movies and TV Shows', fontsize =14)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid lines

plt.legend(title='Type', fontsize=12)  # Add legend with title and adjust font size

plt.tight_layout()
plt.show()
```

```python
# count for yearly increase in the release of shows and movies
# The .size() method in pandas is used to count the number of elements in each group of a DataFrame groupby object. It returns a Series containing the size of each group.
release_year_data = netflix_df.groupby(['release_year', 'type']).size().reset_index(name='count')
release_year_data
```

```python
# Line graph represents the increase in release of movies and TV Shows based on the year

# Set the style of the plot
sns.set_style("whitegrid")

# Create a line plot of release years
plt.figure(figsize=(10, 6))
sns.lineplot(data=release_year_data, x= 'release_year', y ='count', marker='o', hue=release_year_data['type'])

# Set the title and labels
plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Count')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# to creat heatmap for movies and tv shows added in the netflix 

# To get the new df for certain columns 
dateadded_df = netflix_df[['show_id','date_added','country','title']]

# to fetch only month from the date_added column using calendar library
dateadded_df['month_int'] = pd.to_datetime(dateadded_df['date_added']).dt.month
dateadded_df['month'] = dateadded_df['month_int'].apply(lambda x:calendar.month_abbr[x])

# to fetch year from the date_added column
dateadded_df['year'] = pd.to_datetime(dateadded_df['date_added']).dt.year
dateadded_df
```

```python
# creating heatmap df by grouping data based on month and year
heatmap_df = dateadded_df.groupby(['month','year', 'month_int']).size().reset_index(name= 'count').sort_values(['month_int'])

#filter data to get the valid count starting from 2014
heatmap_df = heatmap_df.query('year >= 2014')
heatmap_df
```

```python
# If we try to create a heatmap, it will not work because the data is in a long format
# We have the year and month as indices
# We need the data in a wide format
# ew can do that by using pivot
hp = heatmap_df.pivot(index='year', columns='month', values='count')  # index, columns, values
hp = hp.fillna(0)
hp = hp.astype(int)
hp
```

```python
# Plotting the heatmap 
sns.heatmap(hp, annot=True, fmt='d', cmap='coolwarm')

# Add title and labels
plt.title('Yearly added content Count')
plt.xlabel('Year')
plt.ylabel('Month')
```
